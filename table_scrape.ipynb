```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_countries_by_number_of_billionaires'
# Send a request to the website and retrieve the HTML content
response = urllib.request.urlopen(link)
#retrieve the html content using the below code line 
html_content = response.read()
print(html_content)
```

```python
html_content
```

```python
# create a beautifulsoup object 
soup = BeautifulSoup(html_content,'html.parser')
soup
```

```python
# find the table using html class or tag 
#soup.find() this specifies the first table in the webpage 
table = soup.find('table')
```

```python
FirstTable_data = []
headers = []
# find out all the rows of a table 
rows = soup.find_all('tr')
```

```python
# finding all the headers cells or column names of a table from row[0]
header_row = rows[0]
header_cells = header_row.find_all("th")
```

```python
#We find all the header cells (<th> tags) within the header row using find_all(), and extract their text content using the text attribute. The extracted header values are stored in the headers list.
headers = [cell.text.strip() for cell in header_cells]
```

```python
print(headers)
```

```python
# extracting data from the cells 
for row in rows[1:]:
    cells = row.find_all("td")  # Find all data cells (td)
    row_data = [cell.text.strip() for cell in cells]
    FirstTable_data.append(row_data)
```

```python
print("Table headers:", headers)
print("Table data:")
for row in FirstTable_data:
    print(row)
```
